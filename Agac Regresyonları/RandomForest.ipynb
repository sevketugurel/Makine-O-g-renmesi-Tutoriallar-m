```python
import pandas as pd 
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/sevketugurel/Desktop/Salaries.csv")
veri = data.copy()

print(veri)

y = veri["Salary"]
X = veri["Level"]

y = np.array(y).reshape(-1,1)
X = np.array(X).reshape(-1,1)
```

```python
dtModel = DecisionTreeRegressor(random_state=0)
dtModel.fit(X,y)
dtTahmin = dtModel.predict(X)

rfModel = RandomForestRegressor(random_state=0)
rfModel.fit(X,y)
rfTahmin = rfModel.predict(X)
```

```python
plt.scatter(X,y,color = "red")
plt.plot(X,dtTahmin,c="blue")
plt.plot(X,rfTahmin,c="green")
plt.show()
```
